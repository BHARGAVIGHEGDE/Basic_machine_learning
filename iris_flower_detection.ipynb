IMPORTING LIBRARIES

```python
#importing all the required libraries for this project
import numpy as np
import matplotlib as mp
import seaborn as sb
import pandas as pd
from sklearn import metrics


```

LOADING THE DATASET


```python
columns_name=['sepal length','sepal width','petal length','petal width','flower_class']
#load the data by reading the csv files
data=pd.read_csv('iris.data',names=columns_name)
#listing out all the 150 entries
data.head(150)
```

VISUALIZATION OF DATASET


```python
data.describe()
```

```python
#visualize the whole dataset
#pairplot function is used to plot multiple pairwise bivariate distributions in a dataset
sb.pairplot(data,hue='flower_class')
```

SEPERATING INPUT AND OUTPUT COLUMNS

```python
data_values=data.values
input=data_values[:,0:4]
output=data_values[:,4]
print(input)
print(output)
```

SPLITTING THE DATA INTO TRAINING AND TESTING


```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(input,output,test_size=0.2)
```

TRAINING MODEL USING LOGISTIC REGRESSION ALGORITHM


```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(input_train,output_train)
```

ACCURACY OF MODEL


```python
prediction=model1.predict(input_test)
#accuracy of model
from sklearn.metrics import accuracy_score
print(accuracy_score(output_test,prediction)*100)
```

CLASSIFICATION REPORT

```python
from sklearn.metrics import classification_report
print(classification_report(output_test,prediction))
```

VERIFICATION OF CODE


```python
new=np.array([[3,2,1,0.2],[4.9,2.2,3.8,1.1],[5.3,2.5,4.6,1.9]])
prediction1=model1.predict(new)
print("Prediction of Species: {}".format(prediction1))
```
